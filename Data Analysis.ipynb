```python
read_path = path+'/model_output/traffic/DPG_5v5_v3/statistics.csv'
read_path
```

```python
stats = pd.read_csv(read_path).set_index('episode')
print(stats.shape)
stats.head()
```

```python
#collision for agent0
stats.iloc[::10,:10].sum(axis=1).plot(alpha=0.5);
plt.ylabel('collision');
plt.title('Total collision in an episode');
```

```python
#reward 
stats.iloc[::10,10:20].sum(axis=1).plot(alpha=0.5);
plt.ylabel('reward');
plt.title('Total reward in an episode');
```

```python
for i in range(10,20):
    stats.iloc[::10,i].plot(alpha=0.5);
    plt.legend();
    plt.ylabel('reward');
    plt.title('Reward in an episode');
    plt.show();
```

```python
#loss for agent0
stats.iloc[::10,20:30].mean(axis=1).plot(alpha=0.5);
plt.ylabel('loss');
plt.title('Mean loss (DPG) in an episode');
```
